## Probabilidade Simples


Levando em conta todo o dataset, qual a probabilidade de um jogador que jogou na premier league de 2021-2022 ter 25 anos ou mais?

A ideia desse dado, é entender qua a média de idade entre os jogadores presentes na Premier League e assim determinar a idade que o jogador chega em seu apogeu.

```python
probJogarIdade = len(df[df['Age'] >= 25.0]) / len(df)
print(f'A probabilidade de um jogador que jogou na premier league de 2021-2022 ter 25 anos ou mais é de: {str(probJogarIdade*100)[:6]}%')
```

Nesse caso a idade média segundo algumas pesquisas é de 26 anos, assim confirmando a teoria de que os jogadores da Premier League (liga tão disputada), estão jogando em seu auge.
https://www.90min.com/pt-BR/posts/26-dias-para-a-copa-do-mundo-26-anos-e-a-idade-ideal-para-brilhar-e-ser-campeao-do-torneio#:~:text=Existe%20realmente%20a%20idade%20que,de%20uma%20Copa%20do%20Mundo.

## Probabilidade da União

Qual a probabilidade de um jogador que jogou na premier league de 2021-2022 ter 25 anos ou mais e ser ingles?

```python
#Existe um caracter escondido entre o eng e o ENG
probJogarIdadeANDSerIngles = len(df[(df['Age'] >= 25.0) | (df['Nation'] == 'eng ENG')]) / len(df)
print(f'A probabilidade de um jogador que jogou na premier league de 2021-2022 ter 25 anos ou mais e ser ingles é de: {str(probJogarIdadeANDSerIngles*100)[:6]}%')
```

Partindo desse presuposto, parti em fazer as buscas para validar tamanha porcentagem. Com isso achei esse site: https://www.transfermarkt.com.br/premier-league/gastarbeiter/wettbewerb/GB1/saison_id/2021 , onde somei as porcentagens de jogadores estrageiros na premier league na mesma data, e cheguei a conclusão de que mais de 50% dos jogadores não são ingleses!

Porém duas coisas devem ser levadas em consideração, a primeira sendo a limpeza do dataset, onde removi cerca de 80 valores nulos, o que nesse caso poderia ser referente a jogadores estrangeiros, e segundo, a quantidade de jogadores que são exportados para essa liga é muito grande, assim, tendendo a ter um número proximo de 50%.

## Probabilidade da Intersecção

Qual a probabilidade de ter jogado 20 ou mais jogos e ter feito ao menos 1 gol?

```python
JogarJogosEFazerGols = len(df[(df['Starts'] >= 20) & (df['Gls'] >= 1)])
probJogarJogosEFazerGols = JogarJogosEFazerGols/len(df)

print('A probabilidade de ter jogado 20 ou mais jogos e ter feito ao menos 1 gol é de : '+ str(round(probJogarJogosEFazerGols*100, 3)) + '%')
```

## Probabilidade do Complemento

Qual a probabilidade de ter feito ao menos 1 gol e ter menos de 20 jogos?

```python
ProbJogarVinteJogosOuMenos = len(df[(df['Starts'] < 20)])/len(df)
JogarMenosVinteEFazerGol = ProbJogarVinteJogosOuMenos - probJogarJogosEFazerGols

print('A probabilidade de ter feito ao menos 1 gol e ter menos de 20 jogos é de : '+ str(round(JogarMenosVinteEFazerGol*100, 3)) + '%')
```

## Probabilidade da Diferença


Qual a probabilidade de um jogador ter 25 anos ou mais mas não ter jogado mais de 20 jogos?


```python
a = len(df[df['Age'] >= 25.0]) - len(df[(df['Starts'] < 20) & (df['Age'] >= 25.0)])
probDiferenca = a/len(df)
print(f'A probabilidade de um jogador ter mais de 25 anos mas não ter jogado mais de 20 jogos é de : '+ str(round(probDiferenca*100, 3)) + '%')
```

## Probabilidade Condicional

Qual a probabilidade de um jogador ter feito exatamente 1 na temporada, sabendo que ele tem mais de 20 anos?

```python
gols = len(df[df['Gls'] == 1])/len(df)
idade = len(df[df['Age'] > 20])/len(df)
probCond = gols/idade

print('A probabilidade de um jogador ter feito exatamente 1 na temporada, sabendo que ele tem mais de 20 anos é de : '+ str(round(probCond*100, 3)) + '%')
```

## Teorema de Bayes

Qual a probabilidade de um jogador ter mais de 20 anos tal que fez ao menos 1 gol na temporada?

```python
teoremaBayes = (probCond*idade)/(len(df[df['Gls'] >= 1])/len(df))
print('A probabilidade de ter mais de 20 anos tal que fez ao menos 1 gol? é de : '+ str(round(teoremaBayes*100, 3)) + '%')
```

## Intervalo de Confiança

```python
from sklearn.utils import resample 
from sklearn.metrics import accuracy_score

seed(21)

tamanho_amostra = 100
x = np.random.choice(df_modificado['Age'].values, size = tamanho_amostra, replace=True)
n_iterations = 1000

n_size = int(len(x)) 
medians = list() 

for i in range(n_iterations): 
   s = resample(x, n_samples=n_size); 
   m = np.median(s); 
   medians.append(m) 

plt.hist(medians) 
plt.show() 

alpha = 0.95
p = ((1.0-alpha)/2.0) * 100
lower =  np.percentile(medians, p) 
p = (alpha+((1.0-alpha)/2.0)) * 100
upper =  np.percentile(medians, p) 
  
print(f"\nPara um intervalo de confiança de {alpha*100}% intervalo de confiança está entre {lower} e {upper}")
```
